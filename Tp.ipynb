```python
import pandas as pd

# Charger les données
sales_data = pd.read_excel(r'C:\Users\21269\Desktop\TP PYTHON4DS\Sales.xlsx')

# Vérifier et traiter les valeurs manquantes
sales_data.dropna(inplace=True)

# Supprimer les doublons
sales_data.drop_duplicates(inplace=True)

# Calculer les ventes totales (Total Sales)
sales_data['Total Sales'] = sales_data['Order Quantity'] * sales_data['Unit Selling Price']

# Calculer le bénéfice total (Total Profit)
sales_data['Total Profit'] = (sales_data['Unit Selling Price'] - sales_data['Unit Cost']) * sales_data['Order Quantity']

# Calculer les ventes totales
total_sales = sales_data['Total Sales'].sum()

# Calculer le bénéfice total
total_profit = sales_data['Total Profit'].sum()

# Calculer la quantité totale de commandes
total_order_quantity = sales_data['Order Quantity'].sum()

# Calculer la marge bénéficiaire
profit_margin = total_profit / total_sales

# Afficher les résultats
print(f"Ventes Totales: {total_sales}")
print(f"Bénéfice Total: {total_profit}")
print(f"Quantité Totale de Commandes: {total_order_quantity}")
print(f"Marge Bénéficiaire: {profit_margin:.2%}")
```

```python
import pandas as pd

# Chemin vers le fichier local
file_path = r'C:\Users\21269\Desktop\TP PYTHON4DS\sales.xlsx'

# Charger les données depuis le fichier Excel
sales_data = pd.read_excel(file_path, sheet_name='Sales Orders')
regions_data = pd.read_excel(file_path, sheet_name='Regions')

# Afficher les premières lignes de chaque feuille pour vérifier leur contenu
print("Sales Data Head:")
print(sales_data.head())

print("Regions Data Head:")
print(regions_data.head())
```

```python
# Fusionner les données de vente avec les données de région sur la colonne commune 'Delivery Region Index'
merged_data = pd.merge(sales_data, regions_data, left_on='Delivery Region Index', right_on='Index')

# Vérifier si les colonnes ont été ajoutées correctement
print(merged_data[['Delivery Region Index', 'City', 'Latitude', 'Longitude']].head())
```

```python
print(merged_data.columns)
```

```python
# Calculer les ventes totales (Total Sales) et les bénéfices totaux (Total Profit)
merged_data['Total Sales'] = merged_data['Order Quantity'] * merged_data['Unit Selling Price']
merged_data['Total Profit'] = (merged_data['Unit Selling Price'] - merged_data['Unit Cost']) * merged_data['Order Quantity']
```

```python
import folium
from folium.plugins import HeatMap

# Calculer la latitude et la longitude moyennes
latitude_moyenne = merged_data['Latitude'].mean()
longitude_moyenne = merged_data['Longitude'].mean()

# Créer une carte centrée sur la latitude et la longitude moyennes
m = folium.Map(location=[latitude_moyenne, longitude_moyenne], zoom_start=6)

# Ajouter des points pour les ventes
for _, row in merged_data.iterrows():
    folium.CircleMarker(
        location=[row['Latitude'], row['Longitude']],
        radius=row['Total Sales'] / 1000,  # Ajustez le facteur de mise à l'échelle
        color='blue',
        fill=True,
        fill_color='blue',
        fill_opacity=0.6,
        popup=f"Ville: {row['City']}\nVentes: {row['Total Sales']}\nBénéfice: {row['Total Profit']}\nProduits Vendus: {row['Order Quantity']}"
    ).add_to(m)

# Enregistrer la carte
m.save('map_sales.html')
```
